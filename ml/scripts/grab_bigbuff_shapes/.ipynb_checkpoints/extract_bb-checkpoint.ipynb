```python
import cv2
import numpy as np

# code from https://www.learnopencv.com/read-write-and-display-a-video-using-opencv-cpp-python/

#load our input video
cap = cv2.VideoCapture("wind.mp4")

def get_pixel(event,x,y,flags,param):
    global mouseX,mouseY,frame,lower_bound,upper_bound
    if event == cv2.EVENT_LBUTTONDBLCLK:
        #cv2.circle(img,(x,y),100,(255,0,0),-1)
        lower_bound = np.float32(frame[y,x])
        print(lower_bound)
        mouseX,mouseY = x,y
        
    if event == cv2.EVENT_RBUTTONDBLCLK:
        mouseX,mouseY = x,y
        upper_bound = np.float32(frame[y,x])
        print(upper_bound)

# Check if capturing correctly
if (cap.isOpened()== False):
    print("Error opening video stream or file")
    
#grab the max/min color bounds
bound = [[90, 0, 0], [255,80,80]]

bounds_json = {
    "original": [[90, 0, 0], [255,80,80]]
    "all":[[151, 0, 0], [255, 255, 82]],
    "light_blue": [[215, 189, 0], [255, 255, 150]],
}

#load our windows and callback function
cv2.namedWindow('rgb')
cv2.namedWindow('threshold')
#cv2.setMouseCallback('rgb',get_pixel)

# Read until video is completed
b = 0
n = 0
doVid = True
img = None
bounds_index = 0

while(cap.isOpened()):
    # Capture frame-by-frame
    global frame
    bounds_json["original"] = bound
    
    if doVid:
        ret, frame = cap.read()
    else:
        frame = img
        ret = True
        
    if ret == True:

        # Display the resulting frame
        cv2.imshow('rgb',frame)

        gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        
        #grab all the pixels in our range
        mask     = cv2.inRange(frame,tuple(bound[0]),tuple(bound[1]) )
        mask_rgb = cv2.cvtColor(mask,cv2.COLOR_GRAY2BGR)
        
        #grab the threshold of the image
        #(t, thresh) = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY)
        #cvInRangeS(imgHSV, cvScalar(20, 100, 100), cvScalar(30, 255, 255), imgThreshed)
        cv2.imshow('threshold',mask_rgb&frame)
        img = frame
        
    else:
        doVid = False
    
    #img = old_frame
    # Press Q on keyboard to  exit
    key = cv2.waitKey(10) & 0xFF
    
    if key == ord('q'):
        break
        
    #select color channel
    if key in [ord('1'), ord('2'), ord('3')]:
        n = key-49
        
    #select lower or upper
    elif key == ord('l'):
        b = 0
        
    elif key == ord('u'):
        b = 1
        
    #add or subtract from bound
    elif key == ord('='):
        bound[b][n] += 1
        
    elif key == ord('-'):
        bound[b][n] -= 1
    
    elif key == ord('.'):
        bounds_index+=1
        if bounds_index >= len(bounds_json.keys()):
            bounds_index = 0
        
        bound = bounds_json[list(bounds_json.keys())[bounds_index]]
        
    #correct if pixels are out of bounds
    if bound[b][n] < 0:
        bound[b][n] = 0
    if bound[b][n] > 255:
        bound[b][n] = 255
        

# When everything done, release the video capture object
cap.release()

# Closes all the frames
cv2.destroyAllWindows()
print("BOUNDS:{}".format(bound))
```
